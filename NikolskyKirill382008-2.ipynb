### Построение графиков

###  Подбор многочленов степени 1-4

```python
x = year[37:58]
y = T[37:58]

# настраиваем детали отрисовки графиков
plt.figure(figsize=(16, 9))
plt.title("Аппроксимация данных")
plt.xlabel("Year")
plt.ylabel("ΔT")

plt.scatter(x, y)

legend = []

for d in range(1, 5):
    fp = np.polyfit(x, y, d)
    x1 = np.linspace(x[0], x[20], 50)
    y1 = np.polyval(fp, x1)
    f = np.poly1d(fp)
    print("Многочлен %i степени: " % f.order)
    print(f)
    plt.plot(x1, y1, linewidth=2)
    legend.append("Приближение многочлена %i степени" % f.order)
    
n = len(x)
for i in range(0, n):
    h[i] = int(32*((x[i]-1820)/100)**2)
    
plt.plot(x, h)
legend.append("'Классическое' приближение")
plt.legend(legend, loc="upper left")
plt.grid()
pass
```

### Вывод :  
Приближение многочленов степени 4 более точное, чем приближения многочленов степени 1-3.  
"Классическое" приближение меннее точное, чем приближение многочленов степени 2-4.
